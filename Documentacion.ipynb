# Explicación de todas las variables de los jugadores: 
## Shots 
* *goals*: Goles metidos
* *expectedGoals*: Goles esperados bassados en el numero y calidad de oportunidades
* *totalShots*: Tiros totales
* *expectedGoalsOnTarget*: Como de posible un tiro a puerta se podría convertir en gol, en función de la calidad de la oportunidad, calidad del tiro y desde donde se tiró
* *onTargetScoringAttempt*: Tiros entre los 3 palos (incluidos los goles)
* *shotOffTaget*: Tiros fuera de la porteria
* *hitWoodwork*: Tiros al palo
* *blockedScoringAttempt*: Tiros bloqueados
* *bigChanceMissed*: Grandes ocasiones falladas
* *penaltyMiss*: Penaltis fallados
* *totalOffside*: Fueras de juego

## Pass 
* *goalAssist*: Asistencias
* *expectedAssists*: Asistencias esperadas
* *keyPass*: Pases claves
* *bigChanceCreated*: Grandes ocasiones creadas
* *totalPass*: Pases totales, en principio debería ser el sumatorio de "totalOwnHalfPass" y "totalOppositionHalfPasses", pero no es así en algunos casos
* *accuratePass*: Pases acertados
* *totalOwnHalfPasses*: Pases en campo propio
* *accurateOwnHalfPasses*: Pases en campo propio acertados
* *totalOppositionHalfPasses*: Pases en campo rival totales
* *accurateOppositionHalfPasses*: Pases en campo rival acertados
* *totalLongBalls*: Pases largos totales
* *accurateLongBalles*: Pases largos acertados
* *totalCross*: Centros totales
* *accurateCross*: Centros acertados

## Drib
* *touches*: Toques
* *unsuccessfulTouch*: Toques fallidos
* *totalContest*: Regates
* *wonContest*: Regates ganados
* *wasFouled*: Recibio falta
* *possessionLostCtrl*: Posesiones perdidas
* *penaltyWon*: Penaltis provocados

* *dispossessed*: Le robaron el balón (no está en ninguna categoria)



## Def
* *totalTackle*: Entradas totales
* *wonTackle*: Entradas ganadas
* *interceptionWon*: Intercepciones
* *totalClearences*: Despejes
* *outfielderBlock*: Tiros bloqueados
* *ballRecovery*: Recuperaciones
* *fouls*: Faltas
* *duelWon*: Duelos totales ganados (suma entre los ganados aereos y en tierra)
* *duelLost*: Duelos totales perdidos (suma entre los perdidos aereos y en tierra)
* *aerialWon*: Duelos aereos ganados
* *aerialLost*: Duelos aereos perdidos
* *clearanceOffLine*: Despejes en linea de gol
* *lastManTackle*: Entrada como último hombre
* *challengeLost*: Fue regateado (Creo)
* *errorLeadToAShot*: Errores que llevan a disparo
* *errorLeadToAGoal*: Errores que llevan a gol
* *penaltyConceded*: Penaltis cometidos
* *ownGoals*: Goles en propia

## Gk
* *penaltyFaced*: Penaltis encarados
* *penaltySave*: Penaltis parados
* *saves*: Paradas totales
* *goalsPrevented*: Porcentaje de goles evitados
* *savedShotsFromInsideTheBox*: Paradas a tiros desde dentro del área
* *punches*: Con los puños
* *totalKeeperSweeper*: Salidas del portero
* *accurateKeeperSweeper*: Salidas del portero completas
* *goodHighClaim*: Despejes por alto
* *crossNotClaimed*: Despejes no reclamados

## Rating
* *rating*: Puntuación total en SofaScore
* *ratingVersions.original*: Lo mismo que *rating*
* *ratingVersions.alternative*: Puntuación alternativa
* *minutesPlayed*: Minutos jugados
